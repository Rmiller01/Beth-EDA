```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('melbourne_housing.csv')
```

```python
df.info()
```

```python
property_colors = {
    'H': 'blue',
    'A': 'green',
    'u': 'red',
    't': 'purple'
}
plt.figure(figsize=(10, 6))

for property_type, color in property_colors.items():
    subset = df[df['Type'] == property_type]
    plt.scatter(subset['Landsize'], subset['Price'], label=property_type, color=color, alpha=0.7)

plt.title('Scatter Plot of Land Size vs. Price by Property Type')
plt.xlabel('Land Size')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
average_price_per_year = df.groupby('YearBuilt')['Price'].mean().reset_index()


plt.figure(figsize=(20, 6))
plt.plot(average_price_per_year['YearBuilt'], average_price_per_year['Price'], marker='o', linestyle='-')
plt.title('Average Price per Year Built')
plt.xlabel('Year Built')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
average_price_by_year = df.sort_values(by ='YearBuilt').groupby('YearBuilt',as_index = False).agg({'Price':'mean'})

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year['YearBuilt'], average_price_by_year['Price'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('YearBuilt')
plt.ylabel('Average Price ($)')
plt.title('Average Price per YearBuilt')

# Show the plot
plt.grid(True)
plt.show()
```

```python
average_price_by_region = df.groupby('Regionname')['Price'].mean().reset_index()


plt.figure(figsize=(20, 6))
plt.bar(average_price_by_region['Regionname'], average_price_by_region['Price'])

plt.xlabel('Region')
plt.ylabel('Average Price ($)')
plt.title('Average Property Price by Region')


plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=10)

# Add labels and title
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Property Price Histogram')

# Show the plot
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.boxplot(x='YearBuilt', y='Regionname', data=df, orient='h')
plt.xlabel('Year Built')
plt.ylabel('Region')
plt.title('Horizontal Box Plot of Year Built by Region')
plt.show()
```
